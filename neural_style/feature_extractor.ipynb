```python
#from clip_rn50 import CLIP_RN50
import torch
import clip
import torchvision
from torchsummary import summary
from torchvision import transforms 

from torchvision.models import VGG16_Weights
device = torch.device("cuda")
```

```python
clip.available_models()
model, preprocess = clip.load('RN50x16', device, jit=False)
```

```python
model.visual
```

```python
features = {}
def get_features(name):
    def hook(model, input, output):
        features[name] = output.detach()
    return hook
```

```python
model.visual.layer1.register_forward_hook(get_features('feats1'))
model.visual.layer2.register_forward_hook(get_features('feats2'))
model.visual.layer3.register_forward_hook(get_features('feats3'))
model.visual.layer4.register_forward_hook(get_features('feats4'))
```
